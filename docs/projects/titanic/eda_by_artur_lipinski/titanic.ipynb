## O Danych

Dane o pasażerach Titanica

Zbiór danych zawiera informacje o pasażerach RMS Titanic, który zatonął 15 kwietnia 1912 roku po zderzeniu z górą
lodową. Dane obejmują takie atrybuty jak klasa podróży, wiek, płeć, liczba rodzeństwa/małżonków na pokładzie,
liczba rodziców/dzieci na pokładzie, cena biletu oraz miejsce zaokrętowania.

Zbiór zawiera także informację o tym, czy pasażer przeżył katastrofę.

Titanic przewoził ponad 2,200 osób, z czego ponad 1,500 zginęło, co czyni tę katastrofę jedną z najbardziej
tragicznych w historii morskiej.

Kolumny:

* **pclass** - Klasa biletu
* **survived** - Czy pasażer przeżył katastrofę
* **name** - Imię i nazwisko pasażera
* **sex** - Płeć pasażera
* **age** - Wiek pasażera
* **sibsp** - Liczba rodzeństwa/małżonków na pokładzie
* **parch** - Liczba rodziców/dzieci na pokładzie
* **ticket** - Numer biletu
* **fare** - Cena biletu
* **cabin** - Numer kabiny
* **embarked** - Port, w którym pasażer wszedł na pokład (C = Cherbourg, Q = Queenstown, S = Southampton)
* **boat** - Numer łodzi ratunkowej
* **body** - Numer ciała (jeśli pasażer nie przeżył i ciało zostało odnalezione)
* **home.dest** - Miejsce docelowe

```python
"""
Z powyższej tabeli można zauważyć, że osoby które według danych nie weszły na łodzie ratunkowe, a mimo to przeżyły to 
    głównie kobiety, w tym kilka nastolatek.
A osoby które nie przeżyły mimo że są zarejestrowane że weszły do szalup to głównie mężczyźni w sile wieku od 25 do 36 lat,
    co może pokazywać że te osoby w jakiś sposób oddawali swoje miejsca kobietom, możliwe że wyskakiwali z szalup, 
    aby je ratować. Nie wiemy jak te dane były zbierane, jeśli podczas załadunku do szalup to ta teoria ma sens.
"""
```

```python
# Zobaczy ile osób miało mniej niż 18 lat, ile osób miało mniej niż 10 i 5 lat. Ile z nich weszło do szlup i jaki % przeżyło.


# Filtrujemy osoby według wieku
mniej_18 = df[df["age"] < 18]
mniej_10 = df[df["age"] < 10]
mniej_5 = df[df["age"] < 5]

# Funkcja licząca liczbę osób, które weszły do szalupy oraz przeżywalność
def analiza_wieku(df_group, wiek_opis):
    liczba_osob = len(df_group)
    weszli_do_szalupy = df_group["boat"].notna().sum()
    przezyli = df_group["survived"].sum()
    
    procent_przezycia = (przezyli / liczba_osob) * 100 if liczba_osob > 0 else 0
    
    return {
        "Grupa wiekowa": wiek_opis,
        "Liczba osób": liczba_osob,
        "Weszli do szalupy": weszli_do_szalupy,
        "Przeżyło (%)": round(procent_przezycia, 2)
    }

# Analiza dla każdej grupy wiekowej
wyniki = [
    analiza_wieku(mniej_18, "<18 lat"),
    analiza_wieku(mniej_10, "<10 lat"),
    analiza_wieku(mniej_5, "<5 lat")
]

# Tworzymy DataFrame z wynikami
wyniki_df = pd.DataFrame(wyniki)

# Wyświetlamy tabelę
from IPython.display import display
display(wyniki_df)


```

### Zaczynam sprawdzać korelacje kolumn które mnie interesują

```python
# Sprawdźmy korelacje pomiędzy miejscem wejścia na pokład a przeżywalnością, oraz póżniej uwzględniając to na klasę i płeć.

import seaborn as sns

# Liczymy przeżywalność dla każdego miejsca wejścia na pokład
embarked_survival = df.groupby("embarked")["survived"].mean() * 100

# Wykres przeżywalności dla miejsca wejścia na pokład
plt.figure(figsize=(8, 5))
sns.barplot(x=embarked_survival.index, y=embarked_survival.values, hue=embarked_survival.index, palette="viridis", legend=False)
plt.xlabel("Miejsce wejścia na pokład")
plt.ylabel("Procent przeżywalności (%)")
plt.title("Przeżywalność w zależności od miejsca wejścia na pokład")
plt.ylim(0, 100)
plt.show()






# Liczymy przeżywalność dla każdej klasy biletu (Pclass)
pclass_survival = df.groupby("pclass")["survived"].mean() * 100

# Wykres przeżywalności dla klasy biletu
plt.figure(figsize=(8, 5))
sns.barplot(x=pclass_survival.index.astype(str), y=pclass_survival.values, hue=pclass_survival.index.astype(str), palette="viridis", legend=False)

# Opisy wykresu
plt.xlabel("Klasa biletu")
plt.ylabel("Procent przeżywalności (%)")
plt.title("Przeżywalność w zależności od klasy biletu")
plt.ylim(0, 100)

# Pokaż wykres
plt.show()

```

#### Widać, że osoby z 1 klasy miały 100% większą przeżywalność niż osoby z klasy 3. Ciekawe, że wykres przeżywalności ze względu na wykupioną klase i ze względy na miejsce wejścia na pokład wyglądają podobnie, zobaczmy jak sie korelują ze sobą te dwie wartości.

```python


# Tworzymy kopię oryginalnego DataFrame
df_corr = df.copy()

# One-hot encoding dla klasy biletu (pclass) i miejsca wejścia (embarked)
df_corr = pd.get_dummies(df_corr, columns=["pclass", "embarked"], prefix=["Class", "Embarked"])

# Tworzymy macierz korelacji
correlation_matrix = df_corr[["Class_1.0", "Class_2.0", "Class_3.0", "Embarked_C", "Embarked_Q", "Embarked_S"]].corr()

# Tworzymy heatmapę korelacji
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap="Reds", fmt=".2f", linewidths=0.5, square=True)

# Opisy wykresu
plt.title("Macierz korelacji: Klasa biletu vs Miejsce wejścia")
plt.show()
```
