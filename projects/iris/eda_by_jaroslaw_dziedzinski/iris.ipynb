## **4. Analiza pojedynczych zmiennych**

```python
# Tata potrzebuje teraz podstawowe informacje statystyczne dla każdego z gatunków irysów.
stat_gatunek = df.groupby('klasa (class)', as_index=False).agg(['mean', 'std', 'min', 'max']).round(2)
stat_gatunek
# Po wygenerowaniu statystyk zaobserwował już pewne różnice.
```

## **6. Analiza relacji między zmiennymi**

## TATA aby móc lepiej przyjrzeć się różnicom i relacjom między wartościami zmiennych postanowił zwizualizować na wykresach 

```python
# W związku z tym postanowił przypisać odpowiednim gatunkom irysów - kolory, które pomogą odzwierciedlać uzyskane wyniki relacji na wykresach.

df = df.rename(columns={"klasa (class)": "gatunek"})

# I tak:
## gatunkowi 'Iris-setosa' przypisano kolor nr 10 (odcień niebieskiego),
## gatunkowi 'Iris-versicolor' przypisano kolor nr 20 (odcień zielonego),
## gatunkowi 'Iris-virginica' przypisano kolor nr 30 (odcień czerwonego),

def color_column(gatunek) :
    if gatunek == 'Iris-setosa':
        return 10
    elif gatunek == 'Iris-versicolor':
        return 20
    elif gatunek == 'Iris-virginica':
        return 30
        
df["color_column"] = df["gatunek"].apply(color_column)
df
```

```python
# Obserwując tabelę ze zbiorem danych TATA postanowił wygenerować wykresy z następujacymi relacjami:
df.plot(kind='scatter', x='długość kielicha (sepal length)', y='szerokość kielicha (sepal width)', title="Relacja między długością kielicha a szerokością kielicha", c='color_column', colormap='jet')
df.plot(kind='scatter', x='długość płatka (petal length)', y='szerokość płatka (petal width)', title="Relacja między długością płatka a szerokością płatka", c='color_column', colormap='jet')
df.plot(kind='scatter', x='długość kielicha (sepal length)', y='długość płatka (petal length)', title="Relacja między długością kielicha a długością płatka", c='color_column', colormap='jet')
df.plot(kind='scatter', x='długość kielicha (sepal length)', y='szerokość płatka (petal width)', title="Relacja między długością kielicha a szerokością płatka", c='color_column', colormap='jet')
df.plot(kind='scatter', x='szerokość kielicha (sepal width)', y='szerokość płatka (petal width)', title="Relacja między szerokością kielicha a szerokością płatka", c='color_column', colormap='jet')
df.plot(kind='scatter', x='szerokość kielicha (sepal width)', y='długość płatka (petal length)', title="Relacja między szerokością kielicha a długością płatka", c='color_column', colormap='jet')
```

### Po wygenerowaniu wykresów z relacjami i zależnościami pomiędzy wartościami zmiennych TATA od razu zuważył, że wartości pomiarów jednego gatunku irysów różnią się znacząco od pozostałych dwóch.
### Róźnice w relacji "między długością płatka a jego szerokością" najbardziej pokazują jak poszczególne gatunki różnią sie od siebie i na co zwrócić uwagę przy ich wyborze. Co potwierdziła wizualizacja korelacji między zmiennymi.

```python
# Tata dla pewności wykonał macież korelacyjną 
df[["długość kielicha (sepal length)","szerokość kielicha (sepal width)", "długość płatka (petal length)", "szerokość płatka (petal width)"]].corr()
```

### Najsilniejsza korelacja danych występuje między długością i szerokością platków, zaś najniższa między długością płatka i szerokość kielicha.

### Niemniej jednak analizę trzeba doprowadzić do końca i podsumować całość analizy i wobec tego TATA przystapił do osatniego etapu EDA przed Podsumowaniem.

## **7. Analiza wartości odstających**
