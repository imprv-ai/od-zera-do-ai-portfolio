### A. Kluczowe dane ilustrujące skalę katastrofy

Titanic był w stanie przewozić maksymalnie 3 548 osób, a na pokładzie podczas tragicznego rejsu znajdowało się około 2 200 pasażerów i członków załogi (choć źródła podają różnice rzędu +/- 30 osób).

Statek zachwycał swoją wielkością. Mierzył 269 metrów długości, 28 metrów szerokości i 53 metry wysokości. Dla porównania, zwodowany ponad 20 lat później polski MS Batory miał 160 metrów długości i 22 metry szerokości, a obecnie największy statek pasażerski, Star of the Seas (2025), osiągnie długość 365 metrów, szerokość 66 metrów i pomieści aż 8 000 osób.

Titanic, będący szczytem technologii swoich czasów, zużywał dziennie 825 ton węgla, które zasilały 10 000 żarówek na pokładzie. Na potrzeby rejsu zaopatrzono go w 40 000 świeżych jaj, 34 000 kg mięsa oraz 15 000 butelek piwa. Każdego dnia statek wykorzystywał 53 000 litrów wody pitnej.

*Łodzie ratunkowe – kluczowy problem*

Choć Titanic mógłby pomieścić aż 64 łodzie ratunkowe, pierwotnie zaprojektowano dla niego 48. Ostatecznie liczba ta została zredukowana do zaledwie 20 łodzi, aby zapewnić bardziej estetyczny wygląd pokładów. Te 20 łodzi ratunkowych miało łącznie miejsce dla 1 178 osób – co odpowiadało jedynie jednej trzeciej pasażerów i załogi. W rzeczywistości jednak liczba osób, które faktycznie znalazły się na łodziach, była jeszcze mniejsza.

Przykładowo, pierwsza łódź ratunkowa, mająca pojemność 65 osób, wypłynęła z zaledwie 28 pasażerami na pokładzie. Tymczasem temperatura wody morskiej w miejscu katastrofy wynosiła -2°C, co sprawiło, że szanse przeżycia w wodzie były minimalne. Wrak statku spoczywa obecnie na głębokości 3 800 metrów.

*Szanse przeżycia*

Według danych z Titanic Facts, 32% pasażerów i załogi przeżyło katastrofę. Gdyby jednak wszystkie miejsca w łodziach ratunkowych zostały wykorzystane, przeżywalność mogłaby wynieść około 53%.

*Ciekawostki*

* Katastrofę przeżyły dwa psy, które właściciele zabrali na łodzie ratunkowe.

Jedynie w jednym z materiałów promocyjnych linii White Star Line stwierdzono, że zarówno Olympic, jak i Titanic zostały zaprojektowane tak, aby „jak najbardziej zbliżyć się do niezatapialności”.

Ironią losu jest fakt, że miano „niezatapialnego” przypisano Titanicowi dopiero po katastrofie, kiedy pojawiły się komentarze i wyrazy niedowierzania wobec tak tragicznego w skutkach wydarzenia.

## 4. Analiza wybranych zmiennych

Przeprowadzona analiza opiera się na zbiorze danych reprezentującym 1,310 pasażerów, co stanowi około 60% wszystkich podróżujących (łącznie około 2,200 pasażerów i członków załogi, zgodnie z przywołanymi wcześniej źródłami).

Pytania, na które odpowiada analiza:

* Ile osób przeżyło katastrofę?
* Ile kobiet i mężczyzn podróżowało Titanikiem?
* Kim byli najmłodsi i najstarsi pasażerowie i jaki był średni wiek?

```python
# 1. Obliczam ilu pasażerów przeżyło katastrofę

# Tworzę tabelę Ocaleni
survived_values_count = df['survived'].value_counts()

# Obliczam udział
survived_percentage = round((survived_values_count / total_values) * 100)

# Tworzę tabelę wskazujacą na statystykę Ocalonych
df_survived_count = pd.DataFrame({
    'Ocaleni': survived_values_count,
    'Udział': survived_percentage
    })

df_survived_count.style.set_caption("Ilość i udział Ocalonych")

# Katastrofę przeżyło 500 osób, 38% populacji.
```

```python
# 2. Ile kobiet i mężczyzn podróżowało Titanic'iem?

# Obliczam liczbę podróżujących kobiet i mężczyzn.
gender_values_count = df['sex'].value_counts()

# Obliczam udział płci w grupie pasażerów
gender_percentage = round((gender_values_count / total_values) * 100)

# Tworzę tabelę podziału wg płci 
df_gender_count = pd.DataFrame({
    'Płeć': gender_values_count,
    'Udział': gender_percentage
})

df_gender_count.style.set_caption("Ilość i udział podróżujących ze względu na płeć")

# Na statku przebywały 466 kobiet, które stanowiły 36% populacji.
# W dalszych analizach sprawdzę, czy płeć miała wpływ na przetrwanie.
```

```python
# 3. Kim byli najmłodsi i najstarsi pasażerowie i jaki był średni wiek?
age_agg_df = df.agg({'age': ['min','mean', 'median', 'max']})
age_agg_df.style.set_caption("Przekrój pasażerów ze względu na wiek")

# Najmłodszy pasażer nie miał 2 miesięcy, najstarszy zaś osiągnał 80 lat.
# Średnia i mediana dla danych o wieku wynoszą blisko 30 lat, co wynika poczęści z uzupełnienia danych mediana dokonanego w poprzednich krokach.
# W dalszych analizach sprawdzę, czy wiek miał wpływ na przetrwanie.
```

```python
# 4. Ile osób podróżowało w każdej klasie?

# Obliczam liczbę podróżujących w zależności od klasy biletu
class_values_count = df['pclass'].value_counts()

# Obliczam udział klas w grupie pasażerów
class_percentage = round((class_values_count / total_values) * 100)

# Tworzę tabelę danych wg klas podróżnych
class_count = pd.DataFrame({
    'Klasa': class_values_count,
    'Udział': class_percentage
})

class_count.style.set_caption("Ilość i udział podróżnych ze względu na klasę biletu")

# Najwięcej pasażerów zebrała klasa 3 - 709 osób, co stanowi 54% populacji, klasa pierwsza, co ciekawe zajęła drugie miejsce 25% a klasa 2 - 21%.
# W dalszych analizach sprawdzę, czy klasa biletu miała wpływ na przetrwanie.
```

```python
# 5.  Ile osób podróżowało samotnie, a ilu z rodziną?

# Obliczam liczbę podróżujących w podziale na samotnie podróżujących lub w towarzystwie
accompanied_values_count = df['accompanied'].value_counts()

# Obliczam udział
accompanied_percentage = round((accompanied_values_count / total_values) * 100)

# Tworzę tabelę
accompanied_count = pd.DataFrame({
    'Klasa': accompanied_values_count,
    'Udział': accompanied_percentage
})

accompanied_count.style.set_caption("Ilość i udział podróżujących w towarzystwie")

# 60% pasażerów, to jest 791 osób podróżowało samotnie.
# W dalszych analizach sprawdzę, czy status rodzinny miał wpływ na przetrwanie.
```

```python
# 6. Szansa na przetrwanie wg statusu rodzinnego 

# Wybieram pasażerów wg płci 
male_df = df[df['sex'] == 'male']

# Groupuję wg 'class' i statusu 'accompanied', obliczam % 'survival'   
male_survival_rates = male_df.groupby(['pclass', 'accompanied'])['survived'].agg(['sum', 'count']).reset_index()
male_survival_rates['survival_rate'] = male_survival_rates['sum'] / male_survival_rates['count'] * 100

# Tworzę tabelę przestawna z danymi 'male' 
male_pivot_table = male_survival_rates.pivot(index='pclass', columns='accompanied', values=['sum', 'count', 'survival_rate'])
male_pivot_table.columns = ['_'.join(map(str, col)).strip() for col in male_pivot_table.columns.values]
male_pivot_table.reset_index(inplace=True)

# Zmieniam nazwę kolumn
male_pivot_table.columns = ['Class', 'Male_Survived_Alone', 'Male_Survived_With_Family', 'Male_Total_Alone', 'Male_Total_With_Family', 'Male_Survival_Rate_Alone', 'Male_Survival_Rate_With_Family']

# Grupuję dane  wg 'class',  'family status', then calculating survival rates for all passengers
overall_survival_rates = df.groupby(['pclass', 'accompanied'])['survived'].agg(['sum', 'count']).reset_index()
overall_survival_rates['survival_rate'] = overall_survival_rates['sum'] / overall_survival_rates['count'] * 100

# 6. Tworzę tabelę przestawna
overall_pivot_table = overall_survival_rates.pivot(index='pclass', columns='accompanied', values=['sum', 'count', 'survival_rate'])
overall_pivot_table.columns = ['_'.join(map(str, col)).strip() for col in overall_pivot_table.columns.values]
overall_pivot_table.reset_index(inplace=True)

# 7. Zmieniam nazwy kolumn
overall_pivot_table.columns = ['Class', 'Survived_Alone', 'Survived_With_Family', 'Total_Alone', 'Total_With_Family', 'Survival_Rate_Alone', 'Survival_Rate_With_Family']

# 8. Combining the dataframes (aligning on 'Class' column)
combined_df = pd.merge(overall_pivot_table, male_pivot_table, on='Class', how='inner')

combined_df.style.set_caption("6. Survival rates of passengers with and without family, broken down by class. Survival rates of men with family")
```
